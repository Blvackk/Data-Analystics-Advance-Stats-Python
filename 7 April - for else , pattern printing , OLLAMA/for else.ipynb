# for else

```python
nums = [12,15,18,21,26]

for num in nums:
    if num % 5 == 0:
        print(num)
```

```python
nums = [12,19,18,21,26]   # there is no number divissible by 5 , therfore there is no output

for num in nums:
    if num % 5 == 0:
        print(num)
```

```python
nums = [12,18,21,26]

for num in nums:
    if num % 5 == 0:
        print(num)
        break
else:
        print('Number not found')
        
```
