# for Loop

```python
name  = 'nit'

for i in name:    #i means iteration
    print(i)
```

```python
for i in range(5):
    print(i)
```

```python
# Print the value which is divisible by 5

for i in range(1,51):
    if i%5==0:
        print(i)
```

# Break , Continue & Pass

```python
import keyword
keyword.kwlist
```

```python
for i in range (1,11):
    print(i)
```

```python
for i in range(1,11):
    if i ==6:
        break  #As the condition bcomes TRUE the forward code doesn't function
    print(i)
```

```python
for i in range(1,11):
    if i ==6:
        continue  #As the condition bcomes TRUE then it skips the current iteration and continue rest of the code
    print(i)
```

```python
for i in range(1,11):
    if i ==6:
        pass  #As the condition bcomes TRUE the forward code doesn't function
    print(i)
```

```python
# SKIP THE MULTIPLLES OF 3 & 5 

for i in range(1,51):
    if i%3 == 0 or  i%5 == 0:
        continue
    print(i)
```
