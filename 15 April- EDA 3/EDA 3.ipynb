# EDA TECHNIQUE LETS APPLY

```python
clean_data.isnull().sum()
```

```python
clean_data['Age']
```

```python
# Applying MEAN strategy for cleaning the data

import numpy as np
clean_data['Age'] = clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age'])))
#fillna --> Fill missing value , by default it will apply MEAN STRATEGY
clean_data['Age']
```

```python
clean_data['Exp']
```

```python
clean_data['Exp'] = clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
clean_data['Exp']
```

```python
clean_data
```

```python
clean_data['Location']
```

```python
# FOR CATEGORICAL DATA WE APPLY MODE STATEGY
clean_data['Location'] = clean_data['Location'].fillna(clean_data['Location'].mode()[0])
clean_data['Location']
```

```python
clean_data
```

```python
clean_data.info()
```

# CONVERTING IN-BUILT DATA TYPE INTO USER-DEFINED DATA TYPE

```python

clean_data['Age'] = clean_data['Age'].astype(int)
```

```python
clean_data.info()
```

```python
clean_data['Salary'] = clean_data['Salary'].astype(int)
clean_data['Exp'] = clean_data['Exp'].astype(int)
```

```python
clean_data.info()
```

```python
clean_data['Name'] = clean_data['Name'].astype('category')
clean_data['Domain'] = clean_data['Domain'].astype('category')
clean_data['Location'] = clean_data['Location'].astype('category')
```

```python
clean_data.info()
```
