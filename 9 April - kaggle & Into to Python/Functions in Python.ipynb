# Intro to python functions

```python
def greet():  # ---> without argument
    print("GM")

greet()
greet()
```

```python
def sum(x,y):  # ---> with argument  (X,Y FORMAL ARGUMENT)
    add= x+y
    return add
    
sum(3,4)  # ---> (3,4 ACTUAL ARGUMENT)
```

```python
def sum(x,y):
    add = x+y
    print(add)
    
result = sum(3,4)  
print(result)
print(type(result))
```
